```python
def ind(row):
        r = int(row['Id'][:-5])
        return r
df_base['Id'] = df_base.apply(ind, axis = 1)
df_base = df_base.set_index('Id')   
```

### Create index

```python
dims = df_base.shape[1]
n_cells = 72
# how many cells
quantizer = faiss.IndexFlatL2(dims)
idx_l2 = faiss.IndexIVFFlat(quantizer, dims, n_cells)
```

```python
idx_l2.train(np.ascontiguousarray(df_base.values[:50000, :]).astype('float32'))

idx_l2.add(np.ascontiguousarray(df_base.values).astype('float32'))
```

```python
base_index = {k: v for k, v in enumerate(df_base.index.to_list())}

#del df_base
```

```python
df_train = pd.read_csv(r'C:\\Users\\Admin\\Desktop\\data\\train.csv')#, index_col=0)"
df_train.head()
```

Переведем ответы из срок в числа.

```python
def i_id(row):
    r = int(row['Id'][:-6])
    return r
df_train['Id'] = df_train.apply(i_id, axis = 1)
df_train = df_train.set_index('Id')
```

```python
def indt(row):
    r = int(row['Target'][:-5])
    return r
df_train['Target'] = df_train.apply(indt, axis = 1)
```

```python
targets = df_train["Target"]
df_train.drop("Target", axis=1, inplace=True)
```

```python
vecs, idx = idx_l2.search(np.ascontiguousarray(df_train.values).astype('float32'), 5)
```

```python
acc = 0
for target, el in zip(targets.values.tolist(), idx.tolist()):
    acc += int(target in [base_index[r] for r in el])

print(100 * acc / len(idx))
```

Посчитаем евклидово растояние между векторами тренировочной и векторами выбраннами экспертами из общей базы.

```python
def dis(row):
    e = 0
    for i in row:
        e = e +i**2
    return e**0.5
df['tar'] = df.apply(dis, axis = 1)
target_train = df['tar']
```
